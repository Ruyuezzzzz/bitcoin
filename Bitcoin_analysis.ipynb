```python
#creating a function to check the data when the price was between certain amount 
def check_price_table(minValue,MaxValue):
    price=pricedaily[(pricedaily["Close"].values >=int(minValue))&(pricedaily["Close"].values <=int(MaxValue))]
    return price
```

```python
#creating a function to check whether after the price reached certain value it droped again and such situation happened in how many days
def check_price_drop_days(minValue,MaxValue):
    price=pricedaily[(pricedaily["Close"].values >=int(minValue))&(pricedaily["Close"].values <=int(MaxValue))]
    #print(price)
    priced=pricedaily[(pricedaily.index >= price.index[0])&(pricedaily.index <= price.index[-1])] 
    different_days=priced.shape[0]-price.shape[0]
    print( "After the bitcoin price reached {}, it droped again. And such situation happened \nfor {} days before the price got stable above {}." .format(minValue,different_days,minValue))
    for i in priced.index:
        if i not in price.index:
                print(i)
                
    

check_price_drop_days(1000,2000)
```

```python
check_price_drop_days(2000,3000)
```

```python
#calculate the change rate in percent 
#first row can't be calculated using zero instead
changelist=[0,]
for i in range(1,pmonthly.shape[0]):
            change=pmonthly.Close[i]-pmonthly.Close[i-1]
            change_rate=change/pmonthly.Close[i-1]
            #print(change_rate)
            pct=round(change_rate*100,2)
            changelist.append(pct)
            #print(changelist)


pmonthly=pmonthly.assign(Change  = changelist)

#the largest change rate happened between 2013-10 and 2013-11 which is 187.77%
pmonthly.Change.sort_values()
```
